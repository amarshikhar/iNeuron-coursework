### For else

```python
for i in l:
    print(i)
else:
    print("If for loop is going to complete itself then it will come to else")
```

```python
for i in l:
    if i=="add":
        break
    print(i)
else:#will execute only if for loop is exhausted, it will not execute if it encounters break
    print("check this statement")
```

```python
s="Ishu"
for i in s:
    if i=="k":
        break
    print(i)
else:
    print("dont execute it enless it prints my name")
```

```python
d="sadasas"
for i in d:
    if i=="d":
        break#when d is encountered it breaks the loop
    print(i)
else:
    print("sadasas")#as the loop is broken before getting exhausted, else statement is not executed
```

```python
d="sadasas"
for i in d:
    if i=="d":
        continue#when d is encountered it continues the loop
    print(i)
else:
    print("sadasas")#as the loop is exhausted, else statement is executed
```

```python
n=5
for i in range(0,n):# i for rows
    for j in range(0,i+1):# j for column
        print("*",end="")#end="\n" means nextline, but will only jump once the j loop is completed
    print("\r")#carriage return for nextline, just goes to new line, no space will be observed
```
